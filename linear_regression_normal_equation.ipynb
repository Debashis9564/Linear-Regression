```python
def add_bias_variable(X):
    n = X.shape[0]
    X_b = np.concatenate((np.ones((n,1)),X), axis=1)
    return X_b

import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from mpl_toolkits.mplot3d import Axes3D
import plotly.graph_objs as go
import numpy as np

np.random.seed(0)
X1 = 2 * np.random.rand(100, 1)
# X2 = 3 * np.random.rand(100, 1)
# X = np.c_[X1, X2]
y =4 + 3 * X1 + np.random.randn(100, 1)

# print('X:', X)
# print('y:', y)
# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X1, y, test_size=0.2, random_state=42)

X_train_b = add_bias_variable(X_train)
X_test_b = add_bias_variable(X_test)
x_transpose = np.transpose(X_train_b)
x_transpose_x = np.dot(x_transpose, X_train_b)
x_transpose_y = np.dot(x_transpose,y_train)
theta_normal = np.dot(np.linalg.inv(x_transpose_x), x_transpose_y)

optimized_y_test_normal = [np.dot(X_test_b[i], theta_normal) for i in range(X_test.shape[0])]
plt.scatter(X1,y, label= 'actual')
# plt.scatter(X_test, optimized_y_test, label= 'predicted by gradient decent')
plt.scatter(X_test, optimized_y_test_normal, label= 'predicted by normal equation')
plt.legend()
plt.show()
```
